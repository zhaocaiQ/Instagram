# 출구별 게시글현황에서 2018년도에 급증한 거 확인

```python
import json
import matplotlib.pyplot as plt
#Load in cookie
cookies = json.load(open('2.JSON자료\혜화역1번출구_.json', 'rb'))
#리스트에 담긴 text 하나로 합치기
dates = cookies['date']
```

```python
index_2019=[]
index_2018=[]
index_2017=[]
for i in range(len(dates)):
    if '2019' in dates[i]:
        index_2019.append(i)
    if '2018' in dates[i]:
        index_2018.append(i)
    if '2017' in dates[i]:
        index_2017.append(i)
```

```python
print(index_2019)
```

```python
import re
tags = cookies['hashtag']

station = '#혜화역1번출구'
location = '#혜화역'
for i in tags:
    if station in i:
        i.remove(station)
    if location in i:
        i.remove(location)
print(tags)

tag_list_2019 = []
for i in index_2019:
    tag_list_2019.append(tags[i])
tag_list_2019 = sum(tag_list_2019,[])

tag_list_2018 = []
for i in index_2018:
    tag_list_2018.append(tags[i])
tag_list_2018 = sum(tag_list_2018,[])

tag_list_2017 = []
for i in index_2017:
    tag_list_2017.append(tags[i])
tag_list_2017 = sum(tag_list_2017,[])
```

```python
print(tag_list_2019)
```

```python
from collections import Counter
#해시태그 개수 카운트
tag_count_2019 = Counter(tag_list_2019)
#count로 빈도수 카운트한 단어 for문 돌려서 사전형태로 만들기
tags_2019 = dict()
#빈도가 높은 n개 추출.most_common(n)
tags_list_2019 = tag_count_2019.most_common(20)
for tags, counts in tags_list_2019:
    if len(tags) > 1:
        tags_2019[tags] = counts
del(tags_2019['#대학로'])
        
        
#해시태그 개수 카운트
tag_count_2018 = Counter(tag_list_2018)
#count로 빈도수 카운트한 단어 for문 돌려서 사전형태로 만들기
tags_2018 = dict()
#빈도가 높은 n개 추출.most_common(n)
tags_list_2018 = tag_count_2018.most_common(13)
for tags, counts in tags_list_2018:
    if len(tags) > 1:
        tags_2018[tags] = counts
del(tags_2018['#대학로'])

#해시태그 개수 카운트
tag_count_2017 = Counter(tag_list_2017)
#count로 빈도수 카운트한 단어 for문 돌려서 사전형태로 만들기
tags_2017 = dict()
#빈도가 높은 n개 추출.most_common(n)
tags_list_2017 = tag_count_2017.most_common(20)
for tags, counts in tags_list_2017:
    if len(tags) > 1:
        tags_2017[tags] = counts
del(tags_2017['#대학로'])
```

```python
#한글 폰트 설정
import matplotlib
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="1.font\malgunbd.ttf").get_name()
matplotlib.rc('font', family=font_name)
```

```python

#그래프 사이즈 지정
plt.figure(figsize=(12, 6))
#x,y축 네이밍
plt.xlabel('주요단어')
plt.ylabel('빈도수')
#그래프에 그리드 그리기
# plt.grid(True)
#사전의 값 정렬
Sorted_Dict_Values_tags_2019 = sorted(tags_2019.values(), reverse=True)
#사전의 키 정렬
Sorted_Dict_Keys_tags_2019 = sorted(tags_2019, key=tags_2019.get, reverse=True)

#bar그래프 그리기
plt.bar(range(len(tags_2019)), Sorted_Dict_Values_tags_2019, align='center',color='darkviolet')
plt.xticks(range(len(tags_2019)), list(Sorted_Dict_Keys_tags_2019), rotation='70')
plt.savefig("2019_1번출구_해시태그.png", bbox_inches="tight")
plt.show()
```
