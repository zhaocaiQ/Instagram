# 게시글이 많이 올라오는 날은 홍보용글이 많이 올라와서인 것으로 결론

```python
import json
#Load in cookie
gate = json.load(open('2.JSON자료\혜화역출구전체_.json', 'rb'))
dates = gate["date"] #날짜 데이터 불러오기
hashtags = gate['hashtag'] #해시태그 데이터 불러오기
texts = gate['text'] #게시글 데이터 불러오기
```

# 2021년 05월 24일 분석

```python
#해당 연월일 index찾기
index = []
for i in range(len(dates)):
    rows = dates[i].split('-')
    if rows[0] == '2021':
        if rows[1] == '05':
            if rows[2].split('T')[0] == '24':
                index.append(i)
                
#각 해당 인덱스의 해시태그, 텍스트 추출
hashtag = []
text = []
for i in index:
    hashtag.append(hashtags[i])
    text.append(texts[i])
    
#해시태그 카운팅
from collections import Counter
c_tags = Counter(sum(hashtag,[]))
c_text = Counter(sum(text,[]))

#결과
print(c_tags)
print(c_text)
```
