```python
import json
from tqdm import tqdm
#Load in cookie
gate1 = json.load(open('2.JSON자료\혜화역1번출구_.json', 'rb'))
gate2 = json.load(open('2.JSON자료\혜화역2번출구_.json', 'rb'))
gate3 = json.load(open('2.JSON자료\혜화역3번출구_.json', 'rb'))
gate4 = json.load(open('2.JSON자료\혜화역4번출구_.json', 'rb'))

#날짜 데이터 불러오기
dates1 = gate1["date"]
dates2 = gate2["date"]
dates3 = gate3["date"]
dates4 = gate4["date"]
dates_full = gate_full["date"]

# 시간만 가져오기
hour1 = []
for i in range(len(dates1)):
    hour1.append(dates1[i][-8:-6])
hour2 = []
for i in range(len(dates2)):
    hour2.append(dates2[i][-8:-6])
hour3 = []
for i in range(len(dates3)):
    hour3.append(dates3[i][-8:-6])
hour4 = []
for i in range(len(dates4)):
    hour4.append(dates4[i][-8:-6])
```

```python
from collections import Counter
#시간별 카운팅

count_hours1 = Counter(hour1)
count_hours2 = Counter(hour2)
count_hours3 = Counter(hour3)
count_hours4 = Counter(hour4)
```

```python
# 2시간 주기(1번, 4번출구)
hours1 = count_hours1.keys()
count_2hour1 = {}
for i in range(0, 23, 2):
    total = []
    if i < 10:
        hour = count_hours1["0{}" .format(i)]
        total.append(hour)
    else:
        hour = count_hours1["{}" .format(i)]
        total.append(hour)
    
    count_2hour1["{}-{}" .format(i, i+1)] = sum(total)

hours2 = count_hours2.keys()
count_2hour2 = {}
for i in range(0, 23, 2):
    total = []
    if i < 10:
        hour = count_hours2["0{}" .format(i)]
        total.append(hour)
    else:
        hour = count_hours2["{}" .format(i)]
        total.append(hour)
    
    count_2hour2["{}-{}" .format(i, i+1)] = sum(total)
    
hours3 = count_hours3.keys()
count_2hour3 = {}
for i in range(0, 23, 2):
    total = []
    if i < 10:
        hour = count_hours3["0{}" .format(i)]
        total.append(hour)
    else:
        hour = count_hours3["{}" .format(i)]
        total.append(hour)
    
    count_2hour3["{}-{}" .format(i, i+1)] = sum(total)  
    
hours4 = count_hours4.keys()
count_2hour4 = {}
for i in range(0, 23, 2):
    total = []
    if i < 10:
        hour = count_hours4["0{}" .format(i)]
        total.append(hour)
    else:
        hour = count_hours4["{}" .format(i)]
        total.append(hour)
    
    count_2hour4["{}-{}" .format(i, i+1)] = sum(total)
```
