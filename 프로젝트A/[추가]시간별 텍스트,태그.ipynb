```python
import json
import pandas as pd
with open('2.JSON자료\혜화역1번출구_.json','r') as f:
    outlet=json.load(f)
```

```python
outlet.keys()
```

```python
df=pd.DataFrame(outlet)
```

```python
df=df.drop('img',axis=1)
df=df.drop('id',axis=1)
df=df.drop('like',axis=1)
```

```python
df
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
condition=(df['date'].dt.hour==8) | (df['date'].dt.hour==9)
```

```python
df_=df[condition]
```

```python
df_ #08시와 09시에 올라온 
```

```python
import eunjeon
import re
```

```python
mecab=eunjeon.Mecab()
```

```python
text=[]
hashtag=[]
```

```python
for i in df_['text']:
    for j in i:
        j=re.sub('^[ㄱ-ㅎ가-힣 ]', '',j) #한글추출
        j=mecab.nouns(j)                 #명사추출
        for k in j:
            if len(k)!=0 and len(k)!=1:  #1음절 이하 삭제
                text.append(k)
```

```python
stops= ['혜화역','혜화','대학','출구','그램']
for i in df_['hashtag']:
    for j in i:
        for stop in stops:
            j=re.sub(stop,'',j)
        j=re.sub('^[ㄱ-ㅎ가-힣 ]', '',j)
        j=mecab.nouns(j)
        for k in j:
            if len(k)!=0 and len(k)!=1:
                hashtag.append(k)
```
