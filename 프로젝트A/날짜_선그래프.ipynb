```python
import json
from tqdm import tqdm #진행바 표시
from collections import Counter #개수 카운팅
#Load in cookie
cookies = json.load(open('2.JSON자료\혜화역출구전체_.json', 'rb'))
#날짜 데이터 불러오기
dates = cookies['date']
```

```python
#월만 뽑아내기
date_M = []
for i in tqdm(range(len(dates))):
    date_M.append(dates[i][5:7])
print(date_M)
```

# 월별 게시글 빈도수

```python
#한글 폰트 설정
import matplotlib
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="1.font\\malgunbd.ttf").get_name()
matplotlib.rc('font', family=font_name)
```

```python
#월 카운팅
count_date_M = Counter(date_M)
#key값으로 오름차순 정리
sorted_C = dict(sorted(count_date_M.items()))
print(sorted_C)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(13, 6))
#사전의 값
Sorted_Dict_Values_M = sorted_C.values()
#사전의 키
Sorted_Dict_Keys_M = sorted_C.keys()
plt.plot(Sorted_Dict_Keys_M,Sorted_Dict_Values_M, color='rebeccapurple',marker='o', linestyle='--')
plt.fill_betweenx([400,800], 4, 10, color='lightsteelblue', alpha=0.2)
plt.xlabel('month')
plt.ylabel('contents count')
plt.savefig("월별 게시글 수 추이.png", bbox_inches="tight")
plt.show()
```
